``` There are two type of data structure in panda```

1. Series : which is similar to numpy 1D array . In here in addition it assign default index.

2. DataFrame :It is tabular data structure in which data is laid out in rows and columns format similar to csv, SQL file, but can also be used for higher dimensional data sets. DataFrame object can contain homogenous and hetrogenous values and can thought of logical extension of series data structure. In contrast to series where there is one index a Dataframe has one index for columns  and one index for rows. This allow flexiblity accessing and manipulating data`

````python
years=[1,2,3,4]
f1={1:3,2:44,3:34,4:423}
firm1=pd.Series(f1, index=years)
firm1
````

````python
f2={1:12,2:42,3:32,4:223}
firm2=pd.Series(f2, index=years)
firm2
````

````python
f3={1:52,2:62,3:72,4:123}
firm3=pd.Series(f3, index=years)
firm3
````

````python
df1=pd.DataFrame(columns=['firm1','firm2','firm3'], index=years)
````

````python
df1
````

````python
df1.firm1=firm1
df1.firm2=firm2
df1.firm3=firm3
````

````python
df1
````

````python
dft=df1.T # Transposing  which swap the colums with rows. Transpose is to get better view of table
````

````python
dft
````

````python
del dft[1]
````

````python
dft
````

````python
reindex1 =df1.reindex([5,6,7,8], fill_value=0) # fill value will replace NaN with 0
````

````python
reindex1
````

````python
reindex2 =df1.reindex([5,6,1,8])
````

````python
reindex2
````

````python
reindex3 =df1.reindex([5,6,7,8],method='ffill') # forward fill value just after it found value and next NaN will be filled
````

````python
reindex3
````

````python
reindex4 =df1.reindex([5,6,7,8],method='bfill') # forward fill value just after it found value and next NaN will be filled
````

````python
reindex4
````
