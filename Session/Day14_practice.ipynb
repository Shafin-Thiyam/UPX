```python
woh_2015=pd.read_csv(Dataset+'/2015.csv')
woh_2016=pd.read_csv(Dataset+'/2016.csv')
#woh_2017=pd.read_csv(Dataset+'/2017.csv')
```

```python
woh_2015_top10=woh_2015.head(10)
woh_2016_top10=woh_2016.head(10)
#woh_2017_top10=woh_2017.head(10)
```

```python
x_2015=woh_2015_top10['Happiness Rank']
x_2016=woh_2016_top10['Happiness Rank']
#x_2017=woh_2017_top10['Happiness Rank']
y_2015=woh_2015_top10['Economy (GDP per Capita)']
y_2016=woh_2016_top10['Economy (GDP per Capita)']
#y_2017=woh_2017_top10['Economy (GDP per Capita)']
```

```python
df_compare=pd.DataFrame({'GDP 2015':y_2015,'GDP 2016':y_2016})
```

```python
df_compare
```

```python
style.use('ggplot')
plt.plot(x_2016,y_2016,'blue',label='2016',linewidth=5)
plt.title('World Happiness Report 2015-2016')
plt.xlabel('Happiness Rank')
plt.ylabel('GDP')
plt.legend()
plt.grid(True,color='k')
plt.show();
```

```python
woh_2015_top50=woh_2015.head(50)
woh_2016_top50=woh_2016.head(50)
```

```python
x=woh_2015_top50['Happiness Rank'].groupby(woh_2015_top50['Region']).count()
y=woh_2016_top50['Happiness Rank'].groupby(woh_2016_top50['Region']).count()
x
y
```

```python
regions=['ANZ','CEE','EA','LA','ME','NA','SA','WE']
plt.figure(figsize=(20,6))
plt.subplot(122)
plt.bar(regions,x.values,color='b',xlabel='Region',label='2015',width=0.5)
plt.title('World Happiness Report 2015')
plt.ylabel('Number of countries in top 50')
#plt.xlabel('Region')
plt.legend()
plt.subplot(121)
plt.bar(regions,y.values,color='c',label='2016',width=0.3)
plt.title('World Happiness Report 2016')
plt.ylabel('Number of countries in top 50')
plt.xlabel('Region')
plt.legend()
plt.show()
```

```python
score=woh_2016['Happiness Score']
bins=range(2,9)
plt.hist(score,bins,histtype='bar', color='c', rwidth=.7)
plt.xlabel('Happiness score')
plt.ylabel('Number of countries')
plt.title('Histogram')
plt.show();
```

```python
cols=['c','m','r','deeppink','y','g','sandybrown','b']
x.sort_values().values
plt.pie(x.sort_values().values,labels=regions,colors=cols,startangle=90,shadow=True,autopct='%1.1f%%')
plt.title('Pie Plot')
plt.show();
```

```python
x_2016=woh_2016_top10['Happiness Rank']
y_2016=woh_2016_top10['Family']

plt.scatter(x_2016,y_2016,color='r',label='2016')
plt.title('World Happiness Report 2016')
plt.ylabel('Family')
plt.xlabel('Happiness Rank')
plt.legend()
plt.show()
```

```python
sns.lmplot(x='Happiness Rank',y='Health (Life Expectancy)',data=woh_2015)
plt.show()
```

```python
sns.lmplot(x='Happiness Rank',y='Health (Life Expectancy)',data=woh_2016,fit_reg=False,hue='Region')
```

```python
plt.figure(figsize=(30,10))
dataframe_2016_inter=woh_2016.drop(['Country','Happiness Rank','Lower Confidence Interval','Upper Confidence Interval'],axis=1)
sns.boxplot(x="Region", y="Economy (GDP per Capita)",data = dataframe_2016_inter)
```

```python
# Set theme
sns.set_style('whitegrid')
plt.figure(figsize=(30,10))
# Violin plot
pkmn_type_colors = ['#78C850',  # WE
                    '#F08030',  # NA
                    '#6890F0',  # ANZ
                    '#A8B820',  # ME
                    '#A8A878',  # LA
                    '#A040A0',  # SA
                    '#F8D030',  # CEE
                    '#E0C068',  # EA
                    '#EE99AC',  # SSA
                    '#C03028',  # SA
                    
                   ]
sns.violinplot(x='Region', y='Trust (Government Corruption)', 
               data=woh_2016,palette=pkmn_type_colors)
```

```python
plt.figure(figsize=(15,5))
sns.swarmplot(x=woh_2016['Region'],y=woh_2016['Economy (GDP per Capita)'])
plt.xticks(rotation=90)
```

```python
happiness_score=pd.Series(woh_2016['Happiness Score']>5)
plt.figure(figsize=(26,5))
sns.swarmplot(x=woh_2016['Region'], y=woh_2016['Economy (GDP per Capita)'], hue=happiness_score)
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

```python
plt.figure(figsize=(20,10))
corr =woh_2016.corr()
sns.heatmap(corr,cmap="YlGnBu",annot=True)
plt.show();
```
