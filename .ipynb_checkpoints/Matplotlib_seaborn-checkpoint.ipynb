```python
dataframe_2015=pd.read_csv("2015.csv")
dataframe_2016=pd.read_csv("2016.csv")
#dataframe_2017=pd.read_csv("2017.csv")
dataframe_2015_first_ten_countries=dataframe_2015.head(10)
dataframe_2016_first_ten_countries=dataframe_2016.head(10)

x_2015=dataframe_2015_first_ten_countries['Happiness Rank']
y_2015=dataframe_2015_first_ten_countries['Economy (GDP per Capita)']
x_2016=dataframe_2016_first_ten_countries['Happiness Rank']
y_2016=dataframe_2016_first_ten_countries['Economy (GDP per Capita)']
df=pd.DataFrame({'GDP_2015':y_2015,'GDP_2016':y_2016})
df
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
plt.plot(x_2016,y_2016,'blue',label='2016',linewidth=5)
plt.title('World Happiness Report 2015-2016')
plt.ylabel('GDP')
plt.xlabel('Happiness Rank')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
dataframe_2015_first_fifty_countries=dataframe_2015.head(50)
dataframe_2016_first_fifty_countries=dataframe_2016.head(50)
x=dataframe_2015_first_fifty_countries['Happiness Rank'].groupby(dataframe_2015_first_fifty_countries['Region']).count()
y=dataframe_2016_first_fifty_countries['Happiness Rank'].groupby(dataframe_2016_first_fifty_countries['Region']).count()
x
```

```python
from matplotlib import pyplot as plt
regions=['ANZ','CEE','EA','LA','ME','NA','SA','WE']
plt.figure(figsize=(20,6))
plt.subplot(121)
plt.bar(regions,x.values,color='b',label='2015',width=0.5)
plt.title('World Happiness Report 2015')
plt.ylabel('Number of countries in top 50')
plt.xlabel('Region')
plt.legend()
plt.subplot(122)
plt.bar(regions,y.values,color='c',label='2016',width=0.3)
plt.title('World Happiness Report 2016')
plt.ylabel('Number of countries in top 50')
plt.xlabel('Region')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

score = dataframe['Happiness Score']
bins = [0,1,2,3,4,5,6,7]
plt.hist(score, bins, histtype='bar', rwidth=0.8)
plt.xlabel('Happiness score')
plt.ylabel('Number of countries')
plt.title('Histogram')
plt.show()
```

```python
from bokeh.plotting import figure,show,output_notebook

x_2015=dataframe_2015_first_ten_countries['Happiness Rank']
y_2015=dataframe_2015_first_ten_countries['Family']
x_2016=dataframe_2016_first_ten_countries['Happiness Rank']
y_2016=dataframe_2016_first_ten_countries['Family']

p=figure(width=500, plot_height=450, title='Happiness rank vs Family',
         x_axis_label='Happiness Rank',y_axis_label='Family')

p.circle(x_2015,y_2015,color='firebrick',legend='2015')
p.triangle(x_2016,y_2016,color='navy',legend='2016')

p.multi_line([x_2015,x_2016],[y_2015,y_2016],color=['red','blue'])

output_notebook()
show(p)
```

```python
from bokeh.plotting import figure,show,output_notebook
from bokeh.models import HoverTool, BoxSelectTool #For enabling tools

x_2015=dataframe_2015_first_ten_countries['Happiness Rank']
y_2015=dataframe_2015_first_ten_countries['Family']
x_2016=dataframe_2016_first_ten_countries['Happiness Rank']
y_2016=dataframe_2016_first_ten_countries['Family']

#Add tools
TOOLS = [BoxSelectTool(), HoverTool()]

p=figure(width=500, plot_height=450, title='Happiness rank vs Family',
         x_axis_label='Happiness Rank',y_axis_label='Family',tools=TOOLS)
p.circle(x_2015,y_2015,color='firebrick',legend='2015')
p.triangle(x_2016,y_2016,color='navy',legend='2016')
p.multi_line([x_2015,x_2016],[y_2015,y_2016],color=['red','blue'])
output_notebook()
show(p)
```
